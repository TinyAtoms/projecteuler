```python
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (30,5)
```

```python
data = pd.read_json("results.json").T
```

```python
n = [2000, 4000, 6000, 8000, 10000,
20000, 40000, 60000, 80000, 100000,
200000, 400000, 600000, 800000, 1000000,
2000000, 4000000, 6000000, 8000000, 10000000,
15000000, 20000000, 25000000, 30000000
]
```

```python
# data.loc["erato"]
for test in data.index:
    plt.errorbar(n, data.loc[test], label=test)
plt.ylabel("time(ms)")
plt.xlabel("n")
plt.legend()
plt.xticks(n)
plt.title("time to generate primes smaller than n")
```
